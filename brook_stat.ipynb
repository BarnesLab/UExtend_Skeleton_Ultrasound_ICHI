```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
pd.set_option('display.max_rows', None)
```

```python
brooke_df = pd.read_csv('brooke_df.csv')
df1 = brooke_df.drop_duplicates(subset=['Subj#'], keep='first').reset_index(drop=True)
df1['Group'].value_counts()
```

```python
bdf = brooke_df.dropna(subset=['Brooke'])
bdf = bdf.reset_index(drop=True)
```

```python
bdf
```

```python
bdf['Brooke'].std()
```

```python
from itertools import combinations

df = bdf

# Group by subject and calculate absolute differences between each pair of visits
results = []

for subj, group in df.groupby("Subj#"):
    group_sorted = group.sort_values("Visit #")
    visits = group_sorted["Visit #"].tolist()
    brooke_scores = group_sorted["Brooke"].tolist()
    
    for (i, j) in combinations(range(len(visits)), 2):
        diff = abs(brooke_scores[i] - brooke_scores[j])
        results.append({
            "Subj#": subj,
            "Visit1": visits[i],
            "Visit2": visits[j],
            "Brooke_Diff": diff
        })

result_df = pd.DataFrame(results)
result_df
```

```python
for _, row in brooke_df.iterrows():
    subj = str(row['Subj#']).strip()
    visit_date = str(row['Visit Date']).strip()

    if subj and visit_date and visit_date.lower() != 'nan':
        try:
            visit_date_clean = pd.to_datetime(visit_date).strftime('%m-%d-%y')
        except Exception:
            visit_date_clean = visit_date  # fallback if parsing fails

        path = os.path.join(subj, visit_date_clean)
        os.makedirs(path, exist_ok=True)
```
