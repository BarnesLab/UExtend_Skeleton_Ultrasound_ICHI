```python
def listdir_nohidden(path):
    for f in os.listdir(path):
        if not f.startswith('.'):
            yield f
```

```python
filepath = '/mnt/sdb/arafat/U-Extend/'
```

```python
dirs = sorted([dI for dI in listdir_nohidden(filepath + 'videos_skeleton') if os.path.isdir(os.path.join(filepath + 'videos_skeleton',dI))])
```

```python
dirs
```

```python
import re

# Function to extract timestamps and sort files
def get_start_time(file_name):
    # Extract start time using regex
    match = re.search(r"(\d{2}:\d{2}:\d{2}\.\d{3})", file_name)
    return match.group(1) if match else ""
```

```python
import subprocess

for pid in dirs:
    dir_dates = list(listdir_nohidden(filepath + 'videos_skeleton/' + pid))

    for visit, date in enumerate(dir_dates):
        path = f'{filepath}videos_skeleton/{pid}/{date}'
        jpath = f'{filepath}videos_skeleton/{pid}/{date}/2D'
        savepath = f'{filepath}videos_skeleton/{pid}/{date}/3D'
        
        file_list = list(listdir_nohidden(path))
        jfile_list = list(listdir_nohidden(jpath))
        video_files = [file for file in file_list if "cup_flex" in file and file.endswith(".mp4")]
        json_files = [file for file in jfile_list if "cup_flex" in file and file.endswith(".json")]

        video_files = sorted(video_files, key=get_start_time)
        json_files = sorted(json_files, key=get_start_time)

        for v, j in zip(video_files, json_files):
            vpath = f'{path}/{v}'
            jsonpath = f'{jpath}/{j}'
            print(vpath)

            outfile = '3D_' + v.rsplit('.mp4', 1)[0]

            # Run the shell script with vpath and path
            subprocess.run([
                'python',
                'infer_wild.py',
                '--vid_path',
                vpath,
                '--json_path',
                jsonpath,
                '--out_path',
                savepath,
                '--out_file',
                 outfile
            ])
```
