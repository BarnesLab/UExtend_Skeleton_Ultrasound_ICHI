```python
def match(df, action, Arm):
    m = df[(df['Label'] == action) & (df['Arm'] == Arm)]
    if not m.empty:
        start = m['Start'].values[0]
        end = m['Stop'].values[0]
    else:
        start = np.nan
        end = np.nan

    return start, end
```

```python
label_dir = r"C:\U-Extend\IEEE_Codes\New_LABELS"
label_df = brooke_df[['Subj#','Visit Date']]

for index, row in brooke_df.iterrows():
    subj = str(row['Subj#']).strip()
    visit_date = str(row['Visit Date']).strip()
    visit_date_clean = pd.to_datetime(visit_date).strftime('%m-%d-%y')
    label_path = Path(os.path.join(label_dir, subj, visit_date_clean))
    csv_files = list(label_path.glob("*.csv"))

    if csv_files:
        filename = csv_files[0]
        print(filename)
        df = pd.read_csv(filename)
        df['Label'] = df['Label'].astype(str).str.strip()
        df['Arm'] = df['Arm'].astype(str).str.strip()

        rflex1_s, rflex1_e = match(df, 'CURL_Transport-Flex (Best)', 'RIGHT')
        rflex2_s, rflex2_e = match(df, 'CURL_Transport-Flex', 'RIGHT')
        rext1_s, rext1_e = match(df, 'CURL_Transport-Ext (Best)', 'RIGHT')
        rext2_s, rext2_e = match(df, 'CURL_Transport-Ext', 'RIGHT')

        lflex1_s, lflex1_e = match(df, 'CURL_Transport-Flex (Best)', 'LEFT')
        lflex2_s, lflex2_e = match(df, 'CURL_Transport-Flex', 'LEFT')
        lext1_s, lext1_e = match(df, 'CURL_Transport-Ext (Best)', 'LEFT')
        lext2_s, lext2_e = match(df, 'CURL_Transport-Ext', 'LEFT')

        label_df.loc[index, 'rflex1_s'] = rflex1_s 
        label_df.loc[index, 'rflex1_e'] = rflex1_e
        label_df.loc[index, 'rflex2_s'] = rflex2_s
        label_df.loc[index, 'rflex2_e'] = rflex2_e
        label_df.loc[index, 'rext1_s'] = rext1_s
        label_df.loc[index, 'rext1_e'] = rext1_e
        label_df.loc[index, 'rext2_s'] = rext2_s
        label_df.loc[index, 'rext2_e'] = rext2_e
        label_df.loc[index, 'lflex1_s'] = lflex1_s
        label_df.loc[index, 'lflex1_e'] = lflex1_e
        label_df.loc[index, 'lflex2_s'] = lflex2_s
        label_df.loc[index, 'lflex2_e'] = lflex2_e
        label_df.loc[index, 'lext1_s'] = lext1_s
        label_df.loc[index, 'lext1_e'] = lext1_e
        label_df.loc[index, 'lext2_s'] = lext2_s
        label_df.loc[index, 'lext2_e'] = lext2_e
```

```python
label_df.to_csv('label_df.csv',index=False)
```
